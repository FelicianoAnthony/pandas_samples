```python
import pandas as pd
import numpy as np
from string import ascii_lowercase
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = '/Users/Anthony/Desktop/foursquare/sample.csv'
df = pd.DataFrame.from_csv(f, index_col=None)
locs = pd.DataFrame(df.LOCATION.value_counts()).reset_index()
locs = locs.rename(columns={'index': 'location', 'LOCATION': 'count'})
locs
```

```python
fig = plt.figure(figsize=(11,8))
location_lst = [x for x in range(len(locs))]
plt.bar(location_lst, locs['count'].tolist())
plt.xticks(location_lst, locs['location'].tolist(), rotation=45)
```

```python
g = df.groupby(['LOCATION', 'TIME', "TOTAL_SPEND"])
multi = pd.DataFrame(g.size())
multi.head()
```

```python
multi[multi.index.get_level_values('TIME') == 2008.0]
```

```python
mu = multi[multi.index.get_level_values('TIME') == 2008.0].reset_index()

location_lst = [x for x in range(len(mu))]

fig = plt.figure(figsize=(11,8))

plt.bar(location_lst, mu['TOTAL_SPEND'].tolist())
plt.xticks(location_lst, mu['LOCATION'].tolist(), rotation=45)
```
